### Chargement et aper√ßu des donn√©es

Importer le fichier CSV et afficher les premi√®res lignes, les types de colonnes et les statistiques descriptives.


```python
import pandas as pd

data = pd.read_csv('data.csv')

data.head()

data.info()
data.describe()
```

### Pr√©paration de la colonne temporelle

Cr√©er une copie ind√©pendante du DataFrame et convertir la colonne `time` au format datetime.


```python
# S'assurer que df_pre_traitement est bien une copie ind√©pendante
data = data.copy()

# Convertir la colonne 'time' en datetime
data['time'] = pd.to_datetime(data['time'])

print(f"Type de la colonne 'time' apres conversion : {data['time'].dtype}")
```

### Suppression des doublons exacts

√âliminer les lignes identiques, r√©initialiser l‚Äôindex et sauvegarder le DataFrame nettoy√©.


```python
# üîπ Supprimer les doublons exacts sur toutes les colonnes
df_drop_duplicates = data.drop_duplicates(keep='first')

df_drop_duplicates = df_drop_duplicates.reset_index(drop=True)

df_drop_duplicates.head()

df_drop_duplicates.to_csv('data_drop_duplicates.csv', index=False)

print(f"Nombre de lignes apr√®s suppression des doublons : {len(df_drop_duplicates)}")
```
